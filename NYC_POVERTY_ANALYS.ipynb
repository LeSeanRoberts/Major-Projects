### Introduction

Accurate measurement of poverty is central to both empirical research and policy evaluation, particularly in large urban economies where housing costs, labor market heterogeneity, and public transfer systems interact in complex ways. Traditional poverty metrics, most notably the Official Poverty Measure (OPM), rely on a limited income definition and fail to account for major non-discretionary expenses and in-kind benefits that materially affect household well-being. In response to these limitations, the New York City government developed the NYCgov Poverty Measure, which expands the resource framework to include taxes, refundable credits, public assistance, housing costs, childcare expenses, medical out-of-pocket spending, and nutrition and energy assistance.

NYCgov Poverty Measure microdata providse a uniquely detailed statistical environment for modeling poverty outcomes at the individual and poverty-unit levels. Each observation links demographic attributes, educational attainment, labor supply, and household structure to a fully specified income and expense accounting system. This structure enables analysts to move beyond binary poverty classification and instead model continuous outcomes such as income-to-threshold ratios, poverty gaps, and gap indices. These features are particularly well suited for regression-based and distributional modeling approaches, where both the probability and severity of poverty are of interest. Data from years 2005 to 2018 are aggregated to capture demogrophic and economic variability for developing robust statistical analysis and models. 

From a methodological perspective, the dataset supports a wide range of statistical modeling strategies. Binary outcomes such as official and NYCgov poverty status can be analyzed using logistic or probit models, while continuous outcomes including adjusted income, poverty thresholds, and gap measures lend themselves to linear, generalized linear, or quantile regression frameworks. The presence of poverty-unit identifiers facilitates hierarchical and mixed-effects models that account for within-unit dependence, while survey weights allow for design-consistent population inference. Additionally, the modular structure of income components enables counterfactual simulations in which specific taxes, transfers, or expenses are altered to assess their marginal impact on poverty incidence and depth.

Beyond explanatory modeling, the dataset is also well suited for predictive and policy simulation applications. Machine learning classifiers and regularized regression models can be employed to identify high-risk poverty profiles, while microsimulation techniques can be used to evaluate proposed policy reforms under alternative assumptions. By integrating demographic, labor, and fiscal dimensions within a single coherent framework, the NYCgov Poverty Measure microdata provide a robust empirical basis for both statistical inference and decision-support modeling in urban poverty research.

**The NYCgov Poverty Measure calculates a poverty rate using:**  	

• A **poverty threshold** (referred to as NYCgov threshold) based on 5-year average of national expenditures on food, clothing, shelter and utlities, adjusted for the higher cost of housing in NYC. The threshold, NYCgov_Threshold, is also adjusted for family size.	

• An income measure (referred to as NYCgov income) is estimated using: 	

    — Income from various sources as reported in the American Community Survey (ACS).																						
    — Simulated income taxes to estimate post-tax income and the value of tax credits																						
    — Imputed values of near-cash benefits - SNAP(formerly food stamps), WIC (food aid for women, infants and children), school meals, HEAP (assistance with energy bills for low-income households) and housing assistance - are added to income																						

    — Nondiscretionary costs of work (e.g. communting and childcare) and total out-of-pocket medical spending (including the costs of insurance premiums) are imputed and subtracted from income.																						

    —The NYCgov_Income = (PreTaxIncome_PU – NYCgov_IncomeTax – NYCgov_FICAtax) + (NYCgov_SNAP + NYCgov_WIC + NYCgov_SchoolBreakfast + NYCgov_SchoolLunch) + NYCgov_HEAP + NYCgov_Housing - (NYCgov_Commuting + NYCgov_Childcare + NYCgov_MOOP)																						
• Poverty is determined by the formula:  NYCgov_THRESHOLD < > NYCgov_Income	

    —The NYCgov poverty rate = the percent of the New York City population in poverty

```python
import hashlib

def fast_hash_rows(df):
    # Joining strings in a list comprehension is significantly faster than .apply(axis=1)
    # We use a generator to save memory
    row_strings = (
        "||".join(row) for row in df.astype(str).values
    )
    return [hashlib.md5(s.encode()).hexdigest() for s in row_strings]

# Optimized comparison logic
original_hashes = pd.Series([
    h for df in dfs.values() for h in fast_hash_rows(df)
])
agg_hashes = pd.Series(fast_hash_rows(df_all))

# Value_counts comparison is a clever way to ignore row order!
assert original_hashes.value_counts().equals(agg_hashes.value_counts()), "❌ Mismatch!"
```

Since the counts match perfectly, you are likely looking at a **Transformation Issue** rather than a "Logic Error". Here is what is likely happening:

1. Column Ordering: Your hash function uses "||".join(row). This relies on the columns being in the exact same order. If df_all has columns in a different order than the original DataFrames, the string will be different.

2. Whitespace or Precision: Aggregation often trims strings or rounds floats. Even a single trailing space in one version will break the hash.

```python
def get_clean_hashes(df):
    # 1. Sort columns alphabetically so the order is identical for both
    df_sorted = df.reindex(sorted(df.columns), axis=1)
    
    # 2. Convert to string and hash
    # Using a list comprehension is faster than .apply(axis=1)
    return pd.Series([
        hashlib.md5("||".join(row).encode()).hexdigest() 
        for row in df_sorted.astype(str).values
    ])

# Re-run the hashing on both
original_combined = pd.concat(dfs.values(), ignore_index=True)

print("Hashing original data...")
original_hashes = get_clean_hashes(original_combined)

print("Hashing aggregated data...")
agg_hashes = get_clean_hashes(df_all)

# Final Compare
assert original_hashes.value_counts().equals(agg_hashes.value_counts()), \
       "❌ Row-level mismatch still exists!"

print("✅ Row-level integrity verified! (Columns were just out of order)")
```

**Summary of what went wrong:**

1. The Renaming: Your DIS to DS loop changed the names.

2. The Concat: When you performed the aggregation, Pandas (or your code) likely sorted the columns alphabetically.

3. The Hash: The hash function is "dumb"—it doesn't know what column is which; it just sees a long string. If the columns move, the string changes, and the hash breaks.



Using reindex(sorted(df.columns)) is the "gold standard" for this because it removes the **human factor** and **Pandas' internal sorting logic** from the equation.

**Why this is the safest method:**

1. Ensures Deterministic Output: Hashing is only useful if the input is identical every time. If df_A has columns [Name, Age] and df_B has [Age, Name], they contain the same data, but their "stringified" rows are different. Sorting columns alphabetically (A to Z) ensures that regardless of how the data was joined or renamed, the hashing function sees the data in the exact same sequence.

2. Identifies Renaming Errors: If you renamed DIS to DS in one DataFrame but forgot to do it in the other, reindex won't "fix" the name, but it will cause the hash to fail. This is actually what you want—it alerts you that your schema isn't identical.

3. Aligns the "String Join": When you run "||".join(row), Python just looks at position 0, then position 1, etc. Sorting ensures that position 0 is always the same column (e.g., AGEP) across all datasets.

**The "Golden Rule" for Data Integrity Checks**

Whenever you compare two datasets that have gone through different "lives" (one from a raw file, one from a SQL join, one from a concat), always follow these three steps before hashing:

1. Standardize Column Names: (The DIS to DS fix).

2. Standardize Column Order: (df.reindex(sorted(df.columns), axis=1)).

3. Standardize Data Types: (e.g., ensuring both are float64 or both are int64 so you don't get the 1 vs 1.0 mismatch).

```python
import pandas as pd

def transform_to_numeric(column):
    """
    Cleans string columns and converts them to float or Int64.
    Leaves existing int/float columns alone.
    """
    # Only process object (string) columns
    if column.dtype == 'object':
        # Remove commas and strip whitespace
        cleaned_col = column.astype(str).str.replace(',', '', regex=False).str.strip()
        
        # Convert to numeric, coerced errors become NaN
        numeric_col = pd.to_numeric(cleaned_col, errors='coerce')
        
        # Check if the column consists of whole numbers (integers)
        # We drop NaNs to check the nature of existing data
        non_nulls = numeric_col.dropna()
        if not non_nulls.empty and (non_nulls % 1 == 0).all():
            # Convert to nullable Integer type 'Int64' to allow NaN values
            return numeric_col.astype('Int64')
        else:
            # Keep as float if there are decimals
            return numeric_col
    
    return column

# Apply the transformation to all columns in the dataframe
df_all = df_all.apply(transform_to_numeric)

# Verify the changes
print(df_all.info())
print(df_all.head())

# Optionally save the cleaned data
# df_all.to_csv('df_all_cleaned.csv', index=False)
```

#### Model Specifications

**Poverty Incidence Models:**

Binary poverty status is modeled using a weighted generalized linear model:

$$\text{Pr}(Y_{i}^{(1)}=1) = \text{logit}^-1(\alpha + \mathbf{X}_i\beta + u_{p(i)})$$

where  $\,u_{p(i)}\,\sim\,\mathcal{N}(0 \sigma_{u}^2)\,$ captures unobserved poverty-unit heterogeneity. Estimation uses survey-weighted maximum likelihood (Hosmer et al., 2013).

The basic probability model for logit regression based on the sigmoid function:


The predicted probability $p$ that a binary outcome $Y$ equals 1, given input features $X=(x_1\,...\,,x_k)\,$, is:

$$\pi(X) = P(Y=1|X) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 X_1 + \ldots + \beta_k X_k)}}$$

$P(Y=1|X)$: The probability of the "success" or target outcome, often denoted as $p(x)$. Here $Y=1$ is like "yes" or poverty status "true". 


**Poverty Depth and Severity:**

Conditional on poverty, continuous outcomes are modeled using:

$$Y_{i}^{(3)},\,Y_{i}^{(4)}\,=\,\alpha\,+\,\mathbf{X}_i\beta + u_{p(i)}\,+\,\epsilon_i$$

with heteroskedasticity-robust or quantile-based estimators to accommodate skewed income distributions. Choosing quantile regression because of its customary use.

Mathematical Structure of Quantile Regression:

Since the target 'NYCgov_Pov_Stat' is binary categorical or nominal, with selected features that are continuous, quantile regression is an appropriate model. Quantile regression estimates the conditional $q-$th quantile of a dependent variable y given a set of predictors $X$. Unlike Ordinary Least Squares (OLS), which minimizes the sum of squared residuals to find the mean, Quantile Regression minimizes a weighted sum of absolute residuals. The objective is to find the vector $/beta_q$ that minimizes the "Pinball Loss" (or check function) $\rho_q$:

$$\beta \in \mathbb{R}^p \min_{i=1}^{\sum_n} \rho_q(y_i - x_i^\top \beta)$$

Where the loss function $\rho_q(u)$ is identified as:

$$\rho_q(u) = u(q - \mathbb{I}(u < 0)) = \begin{cases} q \cdot u & \text{if } u \geq 0 \\ (q-1) \cdot u & \text{if } u < 0 \end{cases}$$

In this notation:

1. $q\,\in\,(0,1)$ is the target quantile (e.g., $q\,=\,0.5$ for the median). 

2. $y_i - x_i^\top \beta$ is the residual for the $i-$th observation.

3. $\mathbb{I}$ is the indicator function.

Why Use RobustScaler with Quantile Regression?

While Quantile Regression is inherently "robust" to outliers in the dependent variable (or target) $(y)$ because it uses absolute deviations rather than squared ones, it remains sensitive to outliers in the independent variables (features $X$), often called high-leverage points.


**Distributional and Quantile Models:**

To assess heterogeneity across the income distribution, quantile regression is employed:

$$Q_{\tau}\!\left( Y_i^{(2)} \mid X_i \right)
= \alpha_{\tau} + X_i \beta_{\tau}
$$

for $\,\tau\,\in\,\{0.1,\,0.25\,0.5\}\,$, enabling inference on differential policy sensitivity across poverty depths.

Distributional and Quantile Models: 

To assess heterogeneity across the income distribution, quantile regression is employed:

1. To model the entire income distribution: Unlike Ordinary Least Squares (OLS), which estimates the average effect, quantile regression allows researchers to examine how covariates affect different points (e.g., 10th, 50th, 90th percentiles) of the income distribution, providing a more comprehensive view of income disparities.

2. To analyse heterogeneous effects (returns to education/work): It allows the impact of independent variables (like education, gender, or policy interventions) to vary across different income levels, revealing that some factors may have a stronger impact on low-income earners compared to high-income earners.

3. To handle skewed data and outliers: Income data is often positively skewed with significant outliers. Quantile regression is more robust than mean-based methods, as it reduces the influence of extreme values.

4. To identify "winners" and "losers" from policies: It helps determine if a policy (e.g., minimum wage, tax changes) has different, or even opposite, effects on different segments of the income distribution, rather than just shifting the mean.

5. To account for unobserved heterogeneity: It can model the conditional median or other quantiles to account for unobserved factors (like intrinsic motivation or ability) that are not captured in the data, which is particularly useful for analyzing income inequality. 

### Feature Importance and Feature Selection 

Feature selection is a critical step in machine learning, aiming to improve model performance, reduce overfitting, and enhance interpretability by identifying and retaining only the most relevant variables. Two powerful techniques often employed for this purpose are the Random Forest algorithm's inherent feature importance mechanism and Recursive Feature Elimination (RFE).

**Feature Importance with Random Forest**

The Random Forest, an ensemble learning method that builds multiple decision trees, provides a built-in mechanism for assessing the importance of each feature. During the training process, the algorithm can track how much each feature contributes to reducing impurity (e.g., Gini impurity or entropy) across all the trees. A more sophisticated method often used is "permutation importance". After a forest is trained, the values of a single feature are randomly shuffled, and the resulting decrease in model accuracy or increase in error is measured. A larger drop in performance indicates a higher importance for that specific feature. This method is effective because it directly measures how vital a feature is to the model's predictive capability.

**Feature Selection with Meta-Transformer SelectFromModel**

The SelectFromModel class in Scikit-Learn is a meta-transformer designed for automated feature selection. In the context of high-dimensional datasets or large-scale instances (such as those exceeding 100,000 rows), it serves as an efficient alternative to iterative selection methods. It operates by utilizing the internal importance weights assigned to features by a base estimator—typically a tree-based model like Random Forest or a linear model like Lasso—to identify and retain only the most predictive variables.
SelectFromModel is fundamentally a one-pass selector. Its workflow consists of three primary stages:

1. Importance Assignment: The transformer utilizes an underlying model that has either feature_importances_ (for trees) or coef_ (for linear models) attributes.

2. Thresholding: A numerical cutoff is applied. Any feature with an importance score below this threshold is discarded. This threshold can be a specific value, or a statistical heuristic such as the "mean" or "median" of all importance scores.

3. Transformation: The dataset is reduced to a subset of columns, effectively narrowing the input space for the final predictive model.

**The Advantage of Efficiency with SelectFromModel**

The primary advantage of SelectFromModel is its computational speed. Because it does not require retraining the model for every feature removed, it is significantly more scalable for large datasets. For a dataset with $N$ features, RFE (recursive feature elimination) might require $N$ training cycles; SelectFromModel requires only one. This makes it the preferred choice for initial feature pruning in production environments where training time is a bottleneck.


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectFromModel


# List of targets
targets = ['NYCgov_Pov_Stat', 'NYCgov_REL', 'EST_PovGap', 'EST_PovGapIndex']

for target in targets:
    print(f"\n{'='*60}\nAnalyzing Target: {target}\n{'='*60}")

    # 1. Prepare Data
    # Ensure we drop all target variables from X to avoid data leakage
    X = stats_data.drop(columns=targets)
    y = stats_data[target]

    # CORRECTED: Included both X and y for train test split
    X_train, X_test, y_train, y_test = train_test_split(
        X, 
        y, 
        test_size=0.20, 
        random_state=42, 
        shuffle=True
    )

    # 2. Initialize and Fit Model 
    # n_jobs=-1 uses all available CPU cores
    rf_model = RandomForestRegressor(n_estimators=100, n_jobs=-1, random_state=42)
    
    print(f"Training RandomForest on {len(X_train)} samples...")
    # Updated to fit on training data specifically
    rf_model.fit(X_train, y_train)

    # 3. Calculate Feature Importances
    importances = rf_model.feature_importances_
    feature_importances = pd.DataFrame({
        'Feature': X.columns,
        'Importance': importances
    }).sort_values(by='Importance', ascending=False)

    # 4. Visualization
    top_n = 15 
    plot_data = feature_importances.head(top_n)
    plt.clf() 
    plt.barh(plot_data['Feature'], plot_data['Importance'], color='skyblue', edgecolor='navy', alpha=0.8)
    plt.xlabel('Importance Score')
    plt.ylabel('Feature Name')
    plt.title(f'Top {top_n} Features for {target}')
    plt.gca().invert_yaxis() 
    plt.grid(axis='x', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

    # 5. Fast Feature Selection
    print(f"Selecting top features using SelectFromModel...")
    selector = SelectFromModel(rf_model, max_features=5, prefit=True)
    
    support = selector.get_support()
    selected_features = X.columns[support].tolist()

    print("Top 5 Ranked Features (by importance score):")
    print(feature_importances.head(5))
    
    print("\nSelected Features for downstream modeling:")
    print(selected_features)
```

Out of curiosity will also treat the multiple OLS counterpart. For a sample consisting of $n$ observations and $k$ independent variables, the population linear regression model is given by:

$$Y_i = \beta_0 + \beta_1 X_{1i} + \beta_2 X_{2i} + \cdots + \beta_k X_{ki} + \varepsilon_i$$

where:

$Y_i$ is the dependent variable for observation $i$.

$\beta_0$ is the intercept (constant term).

$\beta_1, \beta_2, \ldots, \beta_k$ are the slope coefficients associated with the independent variables.

$X_{ji}$ denotes the value of the $j$-th independent variable for observation $i$, where $j = 1, 2, \ldots, k$.

$\varepsilon_i$ is the error term, capturing unobserved influences on the dependent variable.

OLS Estimation:

Ordinary Least Squares (OLS) estimation determines the values of the regression coefficients by minimizing the sum of squared residuals. A residual is the difference between the observed value of the dependent variable and the value predicted by the linear regression model.

The OLS estimator chooses the coefficient vector that produces the smallest possible total squared error across all observations. Under standard assumptions (linearity, exogeneity, no perfect multicollinearity, homoskedasticity, and no autocorrelation), the OLS estimator is unbiased and efficient within the class of linear estimators.

The ordinary least squares (OLS) estimator is obtained by minimizing the sum of squared residuals. The residual for observation $i$ is defined as the difference between the observed value and the fitted value from the regression model.

$$\varepsilon_i = Y_i - \hat{Y}_i$$

The OLS estimation problem is:

$$\min_{\beta_0, \beta_1, \ldots, \beta_k} \sum_{i=1}^{n}
\left(
Y_i - \beta_0 - \beta_1 X_{1i} - \beta_2 X_{2i} - \cdots - \beta_k X_{ki}
\right)^2$$

Solving this minimization problem yields the OLS estimators for the regression coefficients.

Assumptions for OLS Estimation:

1. Linearity in parameters

2. Random sampling

3. No perfect multicollinearity

4. Zero conditional mean: $\mathbb{E}[\varepsilon_i \mid \mathbf{X}] = 0$

5. Constant variance (homoskedasticity)

NaN standard errors and divide-by-zero warnings appear because the quantile regression fit is essentially perfectly predicting the median, which makes the density estimate at the residuals zero, causing the standard error calculation to blow up.

The warning:

    1. RuntimeWarning: divide by zero encountered in scalar divide

Output:

    1. Sparsity: inf

    2. std err: nan

Translation:

1. Residuals are all zero or extremely close to zero

2. So the estimated density at residual 0 is 0

3. This makes the standard error calculation divide by zero $\rightarrow$ inf $\rightarrow$ nan

The model is essentially deterministic (based on the coefficients):

    const = 2.522e-07 ≈ 0

    Povunit_Rel = 0.1667

    REL = 1.5

Such suggests the target 'NYCgov_REL' may be a linear combination of your predictors. Say,

    NYCgov_REL ≈ 0.1667 * Povunit_Rel + 1.5 * REL

If that’s exactly true in the data, quantile regression has no uncertainty, so it cannot compute standard errors.

### Why Other Attributes in NYC Poverty Data Are More Insightful Than the Poverty Measure Alone

While the poverty rate is a powerful headline statistic, it only scratches the surface of the economic realities faced by New Yorkers. NYC poverty data includes many attributes beyond the poverty measure—such as household composition, employment status, education, housing costs, race/ethnicity, age, disability status, and neighborhood. These variables provide a deeper, more accurate picture of how poverty affects different communities and why certain households remain economically vulnerable.

First, poverty is not a uniform experience across NYC. Households in different boroughs face dramatically different costs of living, housing pressures, and access to services. A family in Manhattan may earn above the poverty threshold yet struggle due to high rent and childcare costs, while a household in a less expensive borough may still face severe hardship even with a similar income. This geographic context is essential for understanding the true burden of poverty in NYC.

Second, poverty is strongly influenced by employment and education. A single poverty rate cannot capture the difference between long-term poverty caused by unemployment or low wages and temporary poverty caused by sudden shocks like job loss or medical expenses. Attributes such as employment status, industry, and educational attainment allow analysts to identify structural barriers that prevent families from escaping poverty and to design targeted interventions such as job training or educational support.

Third, household characteristics like family size, number of dependents, and disability status significantly shape economic vulnerability. A household with multiple dependents or a disabled member faces higher living costs and may require additional support that is not reflected in a simple poverty measure. These attributes reveal the lived realities of poverty and highlight groups that need specialized policy responses.

Finally, NYC’s poverty landscape is shaped by racial and ethnic disparities. Race and ethnicity variables help identify communities that face systemic inequities in income, employment, and housing. Without these attributes, policymakers risk overlooking populations that experience poverty disproportionately and may be left behind by generalized policy solutions.

In summary, while the poverty rate provides a useful summary, NYC poverty data’s full set of attributes is more informative for understanding the complexity of poverty and designing effective policy. By analyzing the broader context—household structure, employment, education, housing costs, and geography—researchers and policymakers can better identify who is most affected, why poverty persists, and what interventions will be most effective in reducing economic hardship in New York City.
